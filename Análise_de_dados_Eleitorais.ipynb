```python
import basedosdados as bd

# Para carregar o dado direto no pandas
df = bd.read_table(dataset_id='br_poder360_pesquisas',
table_id='microdados',
billing_project_id="project-eleicao2022")
```

```python
df.head()
```

```python
pesq_eleicoes_2022 = df[(df.instituto == 'Datafolha') &
   (df.ano == 2022) &
   (df.cargo == 'presidente') &
   (df.sigla_uf.isnull()) &
   (df.tipo == 'estimulada') &
   ((df.nome_candidato == 'Lula') | (df.nome_candidato == 'Bolsonaro') | (df.nome_candidato == 'Ciro') | (df.nome_candidato == 'Simone Tebet')) &
   (df.descricao_cenario.str.contains('cenário 1')==True)].sort_values(by='data')
```

```python
pesq_eleicoes_2022
```

```python
pd.options.plotting.backend = 'plotly'
```

```python
pesq_eleicoes_2022.plot(x='data', y='percentual', color='nome_candidato',
facet_row='turno')
```
